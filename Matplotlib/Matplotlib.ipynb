```python
tips= sns.load_dataset('tips')
tips.head()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex',style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', style='size',)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='size', style='sex',)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', size='size',)
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', size='size',style='smoker')
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, hue='sex', size='size',style='smoker',col='time')
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
```

```python
sns.catplot(x='day', y='total_bill',data=tips) # catagoricaly data show
```

```python
sns.catplot(x='day', y='total_bill',data=tips, kind='box')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)# It shows a line on a 2 dimensional plane
```

```python
sns.pairplot(tips)
```
