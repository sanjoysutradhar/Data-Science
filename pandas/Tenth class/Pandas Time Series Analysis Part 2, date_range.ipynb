# Pandas Time Series Analysis Part 2: date_range

```python
import pandas as pd
```

```python
df=pd.read_csv("AAPL_Historical_Data_Date_Range.csv",)
df.head()
```

```python
date_rng=pd.date_range(start="02/04/2020",end="02/20/2021",freq="B",name="Date")
```

```python
print(date_rng)
```

```python
df.set_index(date_rng,inplace=True)
```

```python
df.head(15)
```

```python
%matplotlib inline
df.High.plot()
```
