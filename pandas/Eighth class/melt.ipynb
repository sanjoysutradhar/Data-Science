# Melt

Melt is used to transform or reshape data and it is similer to pivot table

```python
import pandas as pd
```

```python
df= pd.read_csv("temperature.csv")
df
```

```python
df1=pd.melt(df,id_vars=["day"],)# id_vars is refer to index axis that you want
df1[df1['variable']=='mumbai']
```

```python
df1=pd.melt(df,id_vars=["day"],var_name="city",value_name="temperature")
```
