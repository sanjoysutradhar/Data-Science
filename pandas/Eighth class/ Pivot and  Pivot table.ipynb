# Pivot and  Pivot table

Pivot allow to reshape or transform dataset

```python
import pandas as pd
```

```python
df= pd.read_csv("weather_by_cities.csv")
```

```python
df
```

```python
df.pivot(index="day",columns="city")
```

```python
df.pivot(index="day",columns="city",values="windspeed")
```

```python
# df.pivot(index="windspeed",columns="city") # need to set index
df
```

```python
df1=df.set_index('windspeed')
df1
```

```python
#df1.pivot(index='windspeed',columns='city')
#df.columns
```

```python
df.pivot_table(index='day',columns='city')
```

```python
df1=df.set_index('city')
df1
```

```python
df1.pivot_table(index='city',columns='day')
```

```python
df
```

```python
df.pivot_table(index='city',columns='day')
```

```python
df.pivot_table(index='city',columns='day',aggfunc="count")
```

```python
df.pivot_table(index='city',columns='day',margins=True)
```
