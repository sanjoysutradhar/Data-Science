# Another reshaping technique(Stack Unstack)

# Reshape dataframe using stack/unstack

```python
import pandas as pd
df = pd.read_excel("stocks.xlsx",header=[0,1])
df
```

```python
df.rename(columns={"Unnamed: 0_level_0":""},inplace=True)
df
```

```python
df.stack(level=1)
```

```python
df_stack=df.stack()
df_stack
```

```python
df_s=df.stack(level=1)
df_s
```

```python
import numpy as np
df_s.replace(to_replace=[np.NaN,pd.NaT],value="",inplace=True)
df_s
```

```python
df_s.rename(columns={"Unnamed: 0_level_0":"Date"},inplace=True)
df_s
```

```python
df_s['Date']=pd.to_datetime(df_s['Date'])
df_s
#df_s.unstack()
```

```python
df_s.unstack(level=1)
```

```python
df_s
```

```python
df_stack.unstack()
```
