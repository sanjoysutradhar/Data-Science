```python
df=pd.read_csv("weather_data.csv")
df
```

```python
new_df = df.replace(-99999,np.NaN)
new_df
```

```python
new_df= df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0',
},np.NaN)
new_df
```

```python
# regex
df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True)
new_df
```

```python
new_df = df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]',
},'',regex=True)
new_df
```

```python

df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
new_df=df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
new_df
```

```python
df['score'].replace({'poor', 'average', 'good', 'exceptional'},{1,2,3,4},inplace=True)
```
